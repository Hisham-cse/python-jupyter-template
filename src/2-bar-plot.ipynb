### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]
colors = ['red', 'green', 'blue', 'orange']

# Create the plot
plt.figure(figsize=(8, 5))
plt.bar(categories,values, color=colors)

# Customize the plot
plt.title('Bar Plot of categories,values')
plt.xlabel('categories')
plt.ylabel('values')
plt.grid(True,axis='y')
# plt.legend()

# Show the plot
plt.show()
```
